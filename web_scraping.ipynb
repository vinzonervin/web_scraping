## Web Scraping Postal Codes in Wikipedia

### Getting the data from Wikipedia

```python
import urllib.request
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "lxml")
soup.title.string
```

Viewing data data that had been scraped.

```python
all_tables=soup.find_all("table")
all_tables
```

Slicing unnecessary fields and focusing on the table which incudes the data.

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

Making a loop to add each data in the table into the dataframe.

```python
A=[]
B=[]
C=[]


for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

Making the head or the column names of the data frame.

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df
```

Removing "\n" and na characters to clean the data frame. Drop rows which are empty or borough rows with not assigned value.

```python
df.dropna(axis=0, inplace=True)
df.drop(df.loc[df['Borough']=='Not assigned\n'].index, inplace=True)
df = df.replace('\n','', regex=True)
df.head()
```

Based on the data, There were 103 Postal codes beginning with letter "M" that are located within the city of Toronto in the province of Ontario in Canada.
